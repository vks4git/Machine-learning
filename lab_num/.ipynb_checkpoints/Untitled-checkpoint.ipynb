# Соревновательные сети

Соревновательная сеть на основе MLP состоит из генерирующей сети без скрытых слоёв и описательной сети с такой же структурой. Сети обучаются по очереди, разделение времени обучения задаётся вручную.
Пример обучения на цифре 1:

```python
%matplotlib inline
from main_gal import evaluate
evaluate(1.0 / 3.0, 20, 3000, 1, save_model = True)
```

Не знаю, почему он показал только последнюю картинку, выводиться должны они все. В любом случае, эволюцию шума до единицы вы можете посмотреть в папке 1. Ещё есть также генерация восьмёрки в папке 8, но для неё не сохранена генеративная модель.
А так как для единицы она сохранена, можно быстро генерировать единицы из чего угодно. Например, из шума, вектора единиц и вектора из 0.42.

```python
%matplotlib inline
from generating_tool import generate
import numpy
generate(1, 3000, fname = "noise")
generate(1, 3000, x = numpy.reshape(numpy.ones(3000), (1, 3000)), fname = "ones")
generate(1, 3000, x = numpy.reshape(numpy.ones(3000), (1, 3000)) * 0.42, fname = "42")
```

Также можно сгенерировать и любую другую цифру, но для этого необходимо сначала обучить модель, вызвав evaluate() с параметром save_model = True.
